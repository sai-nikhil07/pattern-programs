# * patterns

```python
# square pattern

n=5
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
```

```python
# incrasing triangle
n=5
for i in range(n):
    for j in range(i+1):
        print("*", end=" ")
    print()
```

```python
# decreading triangle

n=5
for i in range(n):
    for j in range(i,n):
        print("*", end=" ")
    print()
```

```python
# right angle triangle

n=5
for i in range(n):
    for j in range(i,n):
        print(" ", end=" ")
    for i in range(i+1):
        print("*", end=" ")
    print()
```

```python
# right sidedd downword triangle

n=5

for i in range(n):
    for j in range(i):
        print(" ", end=" ")
    for k in range(i,n):
        print("*", end=" ")
    print()
    
```

```python
# hill pattern

n=5
for i in range(n):
    for j in range(i,n):
        print(" ", end=" ")
    for k in range(i):
        print("*", end=" ")
    for l in range(i+1):
        print("*", end=" ")
    print()
```

```python
# reverse hill
n=5
for i in range(n):
    for j in range(i+1):
        print(" ",end=" ")
    for k in range(i,n):
        print("*", end=" ")
    for l in range(i,n-1):
        print("*", end=" ")
    print()
```

```python
# daimond pattern
n=5

for i in range(n-1):
    for j in range(i,n):
        print(" ", end=" ")
    for i in range(i+1):
        print("*",end=" ")
    for i in range(i):
        print("*", end=" ")
    print()

for i in range(n):
    for j in range(i+1):
        print(" ", end=" ")
    for k in range(i,n):
        print("*",end=" ")
    for i in range(i,n-1):
        print("*", end=" ")
    print()

```

```python
# butterfly pattern

n=5

for i in range(n-1):
    for j in range(i+1):
        print("*", end=" ")
    for k in range(i,n-1):
        print(" ", end=" ")
    for k in range(i,n-1):
        print(" ", end=" ")
    for l in range(i+1):
        print("*", end=" ")
    print()

for i in range(n):
    for j in range(i,n):
        print("*", end=" ")
    for k in range(i):
        print(" ", end=" ")
    for k in range(i):
        print(" ", end=" ")
    for l in range(i,n):
        print("*",end=" ")
    print()
```

```python
# left pascle

n=5

for i in range(n-1):
    for i in range(i+1):
        print("*", end=" ")
    print()

for i in range(n):
    for i in range(i,n):
        print("*", end=" ")
    print()
```

# hallow patterns

```python
# square pattern

n=5

for i in range(n):
    for j in range(n):
        print("*",end=" ")
    print()
```

```python
# square parelell bar pattern

n=5

for i in range(n):
    for j in range(n):
        if j==0 or j==(n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# square plus pattern

n=5

for i in range(n):
    for j in range(n):
        if i==n//2 or j==n//2:
            print("*",end=" ")
        else: print(" ",end=" ")
    print()
```

```python
# cross patternn 

n=5

for i in range(n):
    for j in range(n):
        if i==j or i+j==n-1:
            print("*",end=" ")
        else: print(" ", end=" ")
    
    print()
```

```python
# hallow square
n=5

for i in range(n):
    for j in range(n):
        if j==0 or i==0 or j==(n-1) or i==(n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# hallow increasing triangle

n=5
for i in range(n):
    for j in range(i+1):
        if i==(n-1) or j==0 or j==i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# hollow decreasing triangle

n=5

for i in range(n):
    for j in range(i+1):
        if j==0 or i==n-1 or j==i:
            print("*",end=" ")
        else: print(" ",end=" ")
    print()
```

```python
# halow hill
n=5

for i in range(n):
    for j in range(i,n):
        print(" ",end=" ")
    for j in range(i):
        if i==n-1 or j==0: 
            print("*",end=" ")
        else:
            print(" ",end=" ")
    for j in range(i+1):
        if i==n-1 or j==i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# hallow daimond
n=5

for i in range(n-1):
    for j in range(i,n):
        print(" ", end=" ")
    for j in range(i):
        if j==0:
            print("*",end=" ")
        else: 
            print(" ",end=" ")
    for j in range(i+1):
        if j==i:
            print("*", end=" ")
        else:
            print(" ",end=" ")
    print()

for i in range(n):
    for j in range(i+1):
        print(" ", end=" ")
    for j in range(i,n):
        if i==j:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    for j in range(i,n-1):
        if j==n-2:
            print("*", end=" ")
        else:
            print(" ",end=" ")
    print()

```
